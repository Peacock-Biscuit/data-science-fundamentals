```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0.0,1.0)
plt.xlim()
plt.ylim()
# m = 0.5
y1 = x * np.exp(-x) + (x - 0.5) ** 2 / x  * (1 - np.exp(-x))
# m = 0.57
y2 = x * np.exp(-x) + (x - 0.57) ** 2 / x  * (1 - np.exp(-x))
plt.plot(x, y1, label = "m = 0.5", color = "r")
plt.plot(x, y2, label = "m = 0.57", color = "b")
plt.legend()
plt.xlabel('theta')
plt.ylabel('Risk Function')
plt.show()
```
