The loss function is the LINEX (LINear-EXponential) loss given by

L($\theta$; d) = $\exp$(c(d - $\theta$)) - c(d - $\theta$) - 1; where c is a positive constant.

For c = 0:2; 0:5; 1; plot L($\theta$; d) as a function of d - $\theta$

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-50, 50)
plt.xlim(-50, 50)
plt.ylim(-5, 50)
y1 = np.exp(0.2 * x) -0.2 * x - 1
y2 = np.exp(0.5 * x) -0.5 * x - 1
y3 = np.exp(1 * x) - 1 * x - 1
plt.plot(x, y1, label = "c = 0.2", color = "r")
plt.plot(x, y2, label = "c = 0.5", color = "b")
plt.plot(x, y3, label = "c = 1", color = "g")
plt.legend()
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```
